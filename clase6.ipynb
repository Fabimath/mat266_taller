# SVM

Support vector machine (SVM) es un algoritmo de aprendizaje supervisado que se utiliza en muchos problemas de clasificación y regresión, incluidas aplicaciones médicas de procesamiento de señales, procesamiento del lenguaje natural y reconocimiento de imágenes y voz.

![image info](https://la.mathworks.com/discovery/support-vector-machine/_jcr_content/mainParsys/panel/panelParsys/image_747795111.adapt.full.medium.png/1708406772814.png)

El objetivo del algoritmo SVM es encontrar un hiperplano que separe de la mejor forma posible dos clases diferentes de puntos de datos. “De la mejor forma posible” implica el hiperplano con el margen más amplio entre las dos clases, representado por los signos más y menos en la siguiente figura. El margen se define como la anchura máxima de la región paralela al hiperplano que no tiene puntos de datos interiores. El algoritmo solo puede encontrar este hiperplano en problemas que permiten separación lineal; en la mayoría de los problemas prácticos, el algoritmo maximiza el margen flexible permitiendo un pequeño número de clasificaciones erróneas.

![image](https://la.mathworks.com/discovery/support-vector-machine/_jcr_content/mainParsys/image.adapt.full.medium.jpg/1708406772736.jpg)

```r
# Carga el conjunto de datos iris
data(iris)

# Carga la librería e1071 (si aún no está instalada)
# install.packages("e1071")
library(e1071)
```

```r
head(iris)
plot(iris)
plot(iris$Sepal.Length,iris$Sepal.Width)
```

```r
# Ajusta el modelo SVM
modelo_svm <- svm(Species ~ ., data = iris)

# Predice las clases en el conjunto de datos completo
predicciones <- predict(modelo_svm, iris)

# Calcula la precisión
precision <- sum(predicciones == iris$Species) / nrow(iris)
cat("Precisión del modelo SVM:", precision, "\n")
```

```r
# Carga la librería ggplot2 (si aún no está instalada)
# install.packages("ggplot2")
library(ggplot2)

# Agrega las predicciones al conjunto de datos iris
iris$Predicciones <- predicciones
head(iris)
sum(predicciones != iris$Species)
```

```r
# Gráfico de dispersión con colores según las especies reales y predichas
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Predicciones)) +
  geom_point() +
  labs(title = "Predicciones del modelo SVM para el conjunto de datos Iris",
       x = "Longitud del sépalo",
       y = "Ancho del sépalo") +
  scale_color_manual(values = c("setosa" = "blue", "versicolor" = "green", "virginica" = "red"))
```

```r
# Gráfico de dispersión con colores según las especies reales y predichas
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) +
  geom_point() +
  labs(title = "Predicciones del modelo SVM para el conjunto de datos Iris",
       x = "Longitud del sépalo",
       y = "Ancho del sépalo") +
  scale_color_manual(values = c("setosa" = "blue", "versicolor" = "green", "virginica" = "red"))
```
